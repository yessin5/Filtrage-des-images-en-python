```python
from skimage import data
from skimage.util import random_noise
import matplotlib.pyplot as plt
from scipy import ndimage
import numpy as np
from skimage import filters, feature
```

```python
img = data.camera()
plt.imshow(img)
```

```python
image = random_noise(img, mode = "s&p", amount = 0.1)
```

```python
plt.imshow(image)
```

```python
filtre_moy1 = np.ones((3,3))*1/9
filtre_moy2 = np.ones((5,5))*1/25
filtre_moy3 = np.ones((7,7))*1/49
```

```python
image_moy1 = ndimage.convolve(image, filtre_moy1)
plt.imshow(image_moy1)
```

```python
image_moy2 = ndimage.convolve(image, filtre_moy2)
plt.imshow(image_moy2)
```

```python
image_moy3 = ndimage.convolve(image, filtre_moy3)
plt.imshow(image_moy3)
```

```python
image_median1 = ndimage.median_filter(image, 3)
image_median2 = ndimage.median_filter(image, 7)
```

```python
plt.imshow(image_median1)
```

```python
plt.imshow(image_median2)
```

```python
image_gauss1 = filters.gaussian(image, 1)
image_gauss3 = filters.gaussian(image, 3)
image_gauss5 = filters.gaussian(image, 5)
```

```python
plt.imshow(image_gauss1)
```

```python
plt.imshow(image_gauss3)
```

```python
plt.imshow(image_gauss5)
```
